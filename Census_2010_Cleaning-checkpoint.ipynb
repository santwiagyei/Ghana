```python
def extract_table_name(series):
    text_split = [row.split(':') for row in series]
    col_vals = [item[0] for item in text_split]
    return col_vals
```

```python
def extract_descriptions(series):
    text_split = [row.split(':') for row in series]
    col_vals = [item[1] for item in text_split if len(item)>1]
    col_vals = [item for item in col_vals if len(item)>1]
    return col_vals
```

```python
def formatted_description(series):
    row = series.str.lower().str.replace(" ","").str.replace("'","").str.replace(",","").str.replace("contd","")
    return row.str.replace("(",'').str.replace(")",'')
```

# COMBINE THE DFS 

```python
def combined_tables(dictionary, dict_key):
    if len(dictionary.get(dict_key)) == 1:
        new_df = tables[updated_ctables.get(dict_key)[0]]
        new_df = new_df.dropna(thresh = 7)
        new_df = new_df.dropna(how = 'all', axis=1)
        return new_df
    else:
        new_df = pd.concat([tables[key] for key in dictionary.get(dict_key)])
        new_df = new_df.dropna(thresh = 7)
        new_df = new_df.dropna(how = 'all', axis=1)
        return new_df     
```

```python
def clean_and_set_cols(df):
    df_cols = list(df.iloc[0].values)
    df.columns = df_cols
    new_df = df[df[df_cols[0]] != df_cols[0]]
    new_df = df[df[df_cols[1]] != df_cols[1]]
    new_df = df[df[df_cols[2]] != df_cols[2]]
    return new_df.reset_index(drop = True)
```

```python
combined_ctables = {}
for key in updated_ctables.keys():
    df = combined_tables(updated_ctables, key)
    combined_ctables[key] = clean_and_set_cols(df) 
```

# Changing Column Data Types For Each DataFrame

```python
dataframes_cols = []
for i in range(len(mtd)):
    cols = vars()[df_names[i]].columns
    for item in cols:
        if item not in dataframes_cols:
            dataframes_cols.append(item)
dataframes_cols = [str(col) for col in dataframes_cols]
dataframes_cols
```

```python
cat_cols = [] 
for df in df_names:
    if vars()[df].columns.values[0] not in cat_cols:
        cat_cols.append(str(vars()[df].columns.values[0]))
cat_cols.append('Type')
cat_cols.append('Features')
cat_cols
```

```python
float_cols = []
for item in ['Rate','Per','Ratio','Share','Proportion','Average','HouseholdSize']:
    for col in dataframes_cols:
        if item in col:
            float_cols.append(col)
float_cols
```

```python
int_cols = [col for col in dataframes_cols if col not in cat_cols and col not in float_cols]
int_cols
```

```python
for df in df_names:
    vars()[df] = vars()[df].replace({'-': 0})
```

```python
for df in df_names:
    for i in range(len(vars()[df].columns)):
        if vars()[df].columns.values[i] in int_cols:
            try:
                vars()[df][vars()[df].columns.values[i]] = vars()[df][vars()[df].columns.values[i]].astype(int)
            except:
                pass
        elif vars()[df].columns.values[i] in float_cols:
            try:
                vars()[df][vars()[df].columns.values[i]] = vars()[df][vars()[df].columns.values[i]].astype(float)
            except:
                pass
        else:
            try:
                vars()[df][vars()[df].columns.values[i]] = vars()[df][vars()[df].columns.values[i]].astype('category')
            except:
                pass
```
